# Visualisasi Data

## Area Plot

**Kategori kejahatan kecil di London dari tahun 2008-2016**

```python
minor1.plot(kind='area',
            alpha=0.5,
            stacked=False,
            figsize=(20, 15),
            )

plt.title('Area Plot mengenai Kejahatan Kecil di London pada Tahun 2008 - 2016')
plt.xlabel('Tahun')
plt.ylabel('Total Kasus Kejahatan')
plt.show()
```

**Kesimpulan:**
**Jenis kejahatan kecil yang paling banyak dilakukan adalah Burglary in a Dwelling yaitu dengan jumlah lebih dari 150.000 kejahatan**




## Histograms

**Kejahatan besar yang paling banyak dan paling sedikit terjadi**

```python
count, bin_edges = np.histogram(major, 15)
major.plot(kind ='hist',
           figsize=(20, 15),
           bins=15,
           alpha=0.5,
           xticks=bin_edges,
           )

plt.title('Kejahatan Besar di London Tahun 2008 - 2016')
plt.xlabel('Jumlah Kejahatan')
plt.ylabel('Jarak Tahun')
plt.show()
```

**Kesimpulan: Kejahatan besar yang paling banyak di London tahun 2008-2016 adalah "Theft and Handling" dan yang paling sedikit adalah "Sexual Offences".**

## Bar Chart

**Kasus kejahatan di Wilayah London pada Tahun 2008-2016**

```python
borough = df_1['borough'].value_counts()
barv=borough[0:25]
barv.plot(kind='bar', figsize=(20,10), color='skyblue')

plt.title('Kasus Kejahatan di Wilayah London Tahun 2008 - 2016')
plt.xlabel('Wilayah')
plt.ylabel('Total Kejahatan')

plt.show()
```

**Kesimpulan: Dari 25 wilayah yang diambil, wilayah Croydon adalah wilayah dengan kejahatan paling banyak(>80.000) dan yang paling sedikit adalah Wilayah Barking and Dagenham(<60.000)**

```python
x = df_1["year"].value_counts()
x.columns = ["year"]
y = x.sort_values()
y.plot(kind='barh', figsize=(20,10), color='skyblue')

plt.title('Kasus Kejahatan di City of London Tahun 2008 - 2016')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Kejahatan')

plt.show()
```

**Kesimpulan: Angka kejahatan paling tinggi terjadi pada tahun 2016 dan Angka kejahatan paling rendah adalah pada tahun 2013**

## Pie Chart

**Persentase kejahatan besar di London Tahun 2008-2016**

```python
crimes = df_1['major_category'].value_counts()
major2 = pd.DataFrame(data=crimes.index, columns=["major_category"])
major2['values'] = crimes.values

fig = px.pie(major2, values='values', names='major_category', color_discrete_sequence=px.colors.qualitative.Pastel_r)
fig.show()
```

**Kesimpulan: Persentase kasus kejahatan paling tinggi adalah 30,9% dan persentase paling rendah adalah 0,0447%**

```python
colors_list = ['cyan','lightgreen','khaki','red','coral','teal','violet','navy','darkgrey']
explode_list = [0.1, 0, 0, 0, 0, 0, 0, 0.15, 0.3]
dfpie = df_1.groupby(['major_category']).sum()['value'].sort_values(ascending=False)
dfpie.plot(kind='pie',
           figsize=(15, 6),
           startangle=90,
           autopct='%1.3f%%',
           labels=None,
           colors=colors_list,
           pctdistance=1.12,
           explode=explode_list
           )
plt.title('Persentase Kejahatan Besar di London Tahun 2008 - 2016', y=1.12)
plt.ylabel('')
plt.legend(labels=dfpie.index, loc='upper right')
plt.axis('equal')

plt.show()
```

**Kesimpulan: 3 kejahatan dengan persentase teratas adalah Theft and Handling(30,926%), Violence Against the Person(25,410%), dan Criminal Damage(16,517%)**

## Scatter Plot

```python
scatter = minor1[['Theft From Motor Vehicle','Offensive Weapon']]
scatter = scatter.reset_index()
scatter
```

```python
a = scatter.plot(kind='scatter', x='year', y='Theft From Motor Vehicle', figsize=(10, 6), color='seagreen', label='Theft From Motor Vehicle')
b = scatter.plot(kind='scatter', x='year', y='Offensive Weapon', figsize=(10, 6), color='orange',label='Offensive Weapon', ax=a )

plt.title('Perbandingan antara Theft From Motor Vehicle dan Offensive Weapon di London Tahun 2008 - 2016')
plt.xlabel('Tahun')
plt.ylabel('Total Kasus')

plt.show()
```

**Kesimpulan: Untuk kejahatan Theft From Motor Vehicle memiliki sekitar lebih dari 16.000 kasus setiap tahunnya, sedangkan untuk kejahatan Offensive Weapon hanya memiliki sekitar 2000 sampai 4000 kasus setiap tahunnya.**

## Word Clouds

**Menampilkan Jenis kejahatan minor dan major dalam bentuk Word Clouds**

```python
from wordcloud import WordCloud, STOPWORDS
stopwords=set(STOPWORDS)
```

```python
minor_case = ' '.join(df_1.minor_category)

minor_wc = WordCloud(
    background_color='white',
    max_words=200,
    stopwords=stopwords
    )

minor_wc.generate(minor_case)
```

```python
plt.imshow(minor_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
datamajor = ' '.join(df_1.major_category)
london_bridge = np.array(Image.open('/content/drive/MyDrive/Dataset/london_bridge.png'))
```
